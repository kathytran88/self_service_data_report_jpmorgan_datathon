```python
import pandas as pd 
import numpy as np 
df = pd.read_excel("data.xlsx") 
df.head()
```

```python
df = df.dropna().reset_index(drop=True)
display(df)
```

```python
df.iloc[:, 1:] = (df.iloc[:, 1:]
    .apply(lambda col: pd.to_numeric(
        col.astype(str).str.replace('%', '', regex=False).str.replace(r'[^0-9eE\.\-]+', '', regex=True),
        errors='coerce'))
    .astype(float))
```

```python
display(df) 
```

```python
cap_df = pd.read_excel("child.xlsx")
cap_df["County"] = cap_df["County"].astype(str).str.strip()
cap_df["Facility Name"] = cap_df["Facility Name"].astype(str).str.strip()
cap_df["Total Capacity"] = pd.to_numeric(cap_df["Total Capacity"], errors="coerce")

cap_agg = (cap_df
           .groupby("County", as_index=False)
           .agg({"Facility Name": lambda s: "; ".join(pd.unique(s.dropna())),
                 "Total Capacity": "sum"}))

df["_merge_key"] = np.where(df["region"].astype(str).str.strip().str.upper().eq("NYC"),
                            "NYC",
                            df["region"].astype(str).str.strip())

df = df.merge(cap_agg, left_on="_merge_key", right_on="County", how="left") \
       .drop(columns=["_merge_key", "County"])
```

```python
display(df)
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
display(df)
```

```python
df = df.rename(columns={"Total Capacity": "child_capacity", "Region Display": "region"})
df = df.drop(columns=[c for c in ["facility", "Facility", "Facility Name"] if c in df.columns])
```
